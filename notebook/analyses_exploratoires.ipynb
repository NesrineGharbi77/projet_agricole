```python
# Informations générales sur les fichiers
def display_info(df, name):
    print(f"\n{name} Info:\n")
    print(df.info())
    print(f"\n{name} Missing Values:\n")
    print(df.isnull().sum())
```

```python
# Visualisation des données
sns.set(style="whitegrid")
```

```python
# 1. Visualisation de l'évolution de NDVI dans monitoring_cultures
plt.figure(figsize=(10, 6))
sns.lineplot(data=monitoring_cultures, x="date", y="ndvi", hue="parcelle_id")
plt.title("Évolution de NDVI par parcelle")
plt.xticks(rotation=45)
plt.ylabel("NDVI")
plt.show()
```

```python
# Distribution des types de sol
plt.figure()
sns.countplot(data=sols, x='type_sol')
plt.title('Distribution des types de sol')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# Boxplot des propriétés du sol par type
plt.figure()
sns.boxplot(data=sols, x='type_sol', y='matiere_organique')
plt.title('Matière organique par type de sol')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# 3. Corrélations dans les données météo
plt.figure(figsize=(10, 8))
sns.heatmap(meteo_detaillee.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrélation des variables météo")
plt.show()
```

```python
# 4. Progression des rendements
plt.figure(figsize=(10, 6))
sns.lineplot(data=historique_rendements, x="date", y="progression", hue="parcelle_id")
plt.title("Progression des rendements par parcelle")
plt.xticks(rotation=45)
plt.ylabel("Progression")
plt.show()
```

```python
# Visualisation de l'évolution du NDVI
plt.figure()
sns.lineplot(data=monitoring_cultures, x='date', y='ndvi')
plt.title('Evolution du NDVI dans le temps')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# 5. Analyse des propriétés des sols
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sols, x="matiere_organique", y="capacite_retention_eau", hue="type_sol", size="surface_ha")
plt.title("Matière organique vs. capacité de rétention d'eau")
plt.show()
```

```python
# Analyse météorologique
# Agrégation des données météo par jour
meteo_daily = meteo_detaillee.groupby(meteo_detaillee['date'].dt.date).agg({
    'temperature': ['mean', 'min', 'max'],
    'precipitation': 'sum',
    'humidite': 'mean'
}).reset_index()
```

```python
# Visualisation des températures
plt.figure()
plt.plot(meteo_daily['date'], meteo_daily['temperature']['mean'], label='Moyenne')
plt.fill_between(meteo_daily['date'], 
                meteo_daily['temperature']['min'],
                meteo_daily['temperature']['max'], 
                alpha=0.3, label='Min-Max')
plt.title('Evolution des températures')
plt.xlabel('Date')
plt.ylabel('Température (°C)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# Analyse des rendements
plt.figure()
sns.barplot(data=historique_rendements, x='culture', y='rendement_estime')
plt.title('Rendements estimés par culture')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# Corrélations entre les variables de monitoring
plt.figure()
correlation_matrix = monitoring_cultures[['ndvi', 'lai', 'stress_hydrique', 'biomasse_estimee']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Corrélations entre les indicateurs de monitoring')
plt.tight_layout()
```
